```python
from perlin import Perlin
import matplotlib.pyplot as plt
import numpy as np

perlin = Perlin()

# Generate a 2D grid of Perlin noise
a = np.zeros((100, 100))

for i in range(100):
    for j in range(100):
        a[i, j] = perlin.noise_ex((i+65536)/5, (j+65536)/5, 0, 1)

plt.imshow(a, cmap='gray', interpolation='nearest')
plt.show()
```
